```python

from openai import AsyncAzureOpenAI
import asyncio


endpoint = "https://exf23.cognitiveservices.azure.com/"
subscription_key = "CP2SkRm5XJLYH3Pi8yQWEjdWqnJLbIVM3y1KMsnV4a5LjsAT6qKTJQQJ99CFACYeBjFXJ3w3AAAAACOGqtxD"
api_version = "2024-12-01-preview"

deployment_a = "gpt-5.4-mini"
deployment_b = "gpt-5.4-mini"

client = AsyncAzureOpenAI(
    api_version=api_version,
    azure_endpoint=endpoint,
    api_key=subscription_key,
)


async def call_deployment(deployment_name: str, messages: list[dict]) -> str:
    response = await client.chat.completions.create(
        messages=messages,
        max_completion_tokens=4096,
        model=deployment_name,
    )
    return response.choices[0].message.content


async def run_parallel(user_message: str) -> dict:
    print("\n=== [병렬 연결] A, B 동시 호출 ===")
    print(f"질문: {user_message}\n")

    messages_a = [
        {"role": "system", "content": "당신은 보수적이고 신중한 관점으로 답변하는 어시스턴트입니다."},
        {"role": "user", "content": user_message},
    ]
    messages_b = [
        {"role": "system", "content": "당신은 창의적이고 도전적인 관점으로 답변하는 어시스턴트입니다."},
        {"role": "user", "content": user_message},
    ]

    answer_a, answer_b = await asyncio.gather(
        call_deployment(deployment_a, messages_a),
        call_deployment(deployment_b, messages_b),
    )

    print(f"🅰️  답변 A (신중한 관점):\n{answer_a}")
    print(f"\n🅱️  답변 B (창의적 관점):\n{answer_b}")

    return {"answer_a": answer_a, "answer_b": answer_b}


async def run_serial(user_message: str) -> dict:
    print("\n=== [직렬 연결] A → B → 최종 답변(C) ===")
    print(f"질문: {user_message}\n")

    messages_a = [
        {
            "role": "system",
            "content": "당신은 주제를 분석하고 핵심 아이디어 초안을 작성하는 어시스턴트입니다. 간결한 bullet point로 답하세요.",
        },
        {"role": "user", "content": user_message},
    ]
    draft_a = await call_deployment(deployment_a, messages_a)
    print(f"1️⃣ A의 초안:\n{draft_a}")

    messages_b = [
        {
            "role": "system",
            "content": "당신은 전달받은 초안을 검토, 보강하여 완성된 형태의 답변으로 다듬는 편집자입니다.",
        },
        {
            "role": "user",
            "content": (
                f"다음은 원래 질문과 1차 초안입니다.\n\n"
                f"[원래 질문]\n{user_message}\n\n"
                f"[1차 초안]\n{draft_a}\n\n"
                f"이 초안을 다듬고 보강해서 최종 답변을 작성해주세요."
            ),
        },
    ]
    final_c = await call_deployment(deployment_b, messages_b)
    print(f"\n2️⃣ B가 보강한 최종 답변 (C):\n{final_c}")

    return {"draft_a": draft_a, "final_c": final_c}



result_parallel = await run_parallel("재택근무의 장단점을 알려주세요.")
result_serial = await run_serial("재택근무의 장단점을 알려주세요.")
```

```python
# 샘플 3. 역할(Role) 부여 및 프롬프트 엔지니어링

import json
from openai import AzureOpenAI

# 🔧 본인의 Azure OpenAI 정보를 직접 입력하세요
endpoint = "https://exf23.cognitiveservices.azure.com/"
deployment = "gpt-5.4-mini"
subscription_key = "CP2SkRm5XJLYH3Pi8yQWEjdWqnJLbIVM3y1KMsnV4a5LjsAT6qKTJQQJ99CFACYeBjFXJ3w3AAAAACOGqtxD"
api_version = "2024-12-01-preview"

client = AzureOpenAI(
    api_version=api_version,
    azure_endpoint=endpoint,
    api_key=subscription_key,
)


def call_azure_openai(messages, max_completion_tokens=4096, response_format=None):
    kwargs = {
        "messages": messages,
        "max_completion_tokens": max_completion_tokens,
        "model": deployment,
    }
    if response_format:
        kwargs["response_format"] = response_format

    response = client.chat.completions.create(**kwargs)
    return response.choices[0].message.content


# ── [셀 3] 1) 역할 부여: 같은 질문, 다른 페르소나 ────────────
def demo_role(user_message: str) -> None:
    print("\n=== [역할 부여] 같은 질문, 다른 역할(persona) ===")

    personas = [
        {
            "name": "초등학교 선생님",
            "system": "당신은 초등학생에게 설명하는 친절한 선생님입니다. 쉬운 단어와 비유를 사용해 설명하세요.",
        },
        {
            "name": "재무 전문가",
            "system": "당신은 20년 경력의 재무 컨설턴트입니다. 숫자와 리스크 관점에서 전문적으로 답변하세요.",
        },
        {
            "name": "비판적 평론가",
            "system": "당신은 어떤 주장이든 약점과 반론을 먼저 찾아내는 비판적 평론가입니다.",
        },
    ]

    for persona in personas:
        answer = call_azure_openai(
            [
                {"role": "system", "content": persona["system"]},
                {"role": "user", "content": user_message},
            ]
        )
        print(f"\n👤 [{persona['name']}]\n{answer}")


def demo_few_shot(user_message: str) -> None:
    print("\n=== [Few-shot] 예시 기반 프롬프트 엔지니어링 ===")

    messages = [
        {
            "role": "system",
            "content": (
                "당신은 입력된 문장의 감정을 \"긍정/부정/중립\" 중 하나로 분류하고 "
                "이유를 한 줄로 설명하는 분류기입니다. 아래 예시와 동일한 형식으로 답하세요."
            ),
        },
        {"role": "user", "content": "오늘 날씨가 너무 좋아서 기분이 최고예요."},
        {"role": "assistant", "content": "분류: 긍정\n이유: \"기분이 최고\"라는 직접적인 긍정 표현이 있음."},
        {"role": "user", "content": "버스를 놓쳐서 약속에 늦었다."},
        {"role": "assistant", "content": "분류: 부정\n이유: 버스를 놓치고 약속에 늦은 부정적 상황 묘사."},
        {"role": "user", "content": user_message},
    ]

    answer = call_azure_openai(messages)
    print(f"\n📋 결과:\n{answer}")


def demo_json_output(user_message: str) -> None:
    print("\n=== [JSON 강제 출력] 구조화된 응답 ===")

    messages = [
        {
            "role": "system",
            "content": (
                "당신은 사용자의 요청을 분석해서 반드시 아래 JSON 스키마로만 응답하는 어시스턴트입니다. "
                "다른 설명, 마크다운 코드블록 없이 순수 JSON만 출력하세요.\n"
                '{\n  "topic": "string",\n  "summary": "string (2문장 이내)",\n'
                '  "keywords": ["string", "..."],\n  "sentiment": "긍정 | 부정 | 중립"\n}'
            ),
        },
        {"role": "user", "content": user_message},
    ]

    answer = call_azure_openai(messages, response_format={"type": "json_object"})

    print(f"\n🧩 원본 응답 문자열:\n{answer}")

    try:
        parsed = json.loads(answer)
        print(f"\n✅ 파싱된 JSON 객체:\n{parsed}")
    except json.JSONDecodeError:
        print("\n⚠️ JSON 파싱 실패. 모델/배포가 json_object를 지원하는지 확인하세요.")


def demo_chain_of_thought(user_message: str) -> None:
    print("\n=== [Chain-of-Thought] 단계별 사고 유도 ===")

    messages = [
        {
            "role": "system",
            "content": (
                "당신은 문제를 풀 때 반드시 다음 형식을 따르는 어시스턴트입니다:\n"
                "1. [분석] 문제를 단계별로 분석한다.\n"
                "2. [계산/추론] 필요한 계산이나 추론 과정을 보여준다.\n"
                "3. [결론] 최종 결론만 한 줄로 명확히 제시한다."
            ),
        },
        {"role": "user", "content": user_message},
    ]

    answer = call_azure_openai(messages)
    print(f"\n🧠 단계별 답변:\n{answer}")



#demo_role("주식 투자를 처음 시작하려는데 어떻게 해야 하나요?")
#demo_few_shot("드디어 프로젝트가 끝났다. 너무 지친다.")
demo_json_output("최근 전기차 시장 동향에 대해 알려주세요.")
# demo_chain_of_thought("연봉 4000만원에서 연봉 4800만원으로 이직하면 월 실수령액은 대략 얼마나 늘어날까?")
```
